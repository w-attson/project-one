```python
import pandas as pd
from pathlib import Path
```

```python
fatalities_df = pd.read_excel(Path("ardd_fatalities_jul2023_updated.xlsx"), sheet_name="BITRE_Fatality",header=4)
```

```python
fatalities_df['Gender'].head(10)
```

```python
vic_fatalaties_df = fatalities_df[fatalities_df["State"]=='Vic']
# vic_fatalaties_df.count()
```

```python
# vic_fatalaties_df = vic_fatalaties_df.drop(columns=['Bus Involvement','Heavy Rigid Truck Involvement','Articulated Truck Involvement'])
vic_fatalaties_df
```

```python
vic_fatalaties_df.reset_index()
```

```python
vic_fatalaties_df = vic_fatalaties_df.replace({-9:'na','-9':'na'})
vic_fatalaties_df = vic_fatalaties_df.replace({'Monday':'Mon','Tuesday':'Tue','Wednesday':'Wed','Thursday':'Thur','Friday':'Fri','Saturday':'Sat','Sunday':'Sun'})
```

```python
vic_fatalaties_df['Crash ID'] = vic_fatalaties_df['Crash ID'].astype('str')
vic_fatalaties_df.dtypes
```

```python
weekdays = [ 'Mon', 'Tue', 'Wed', 'Thur', 'Fri', 'Sat', 'Sun']

speed = vic_fatalaties_df[vic_fatalaties_df['Speed Limit']!='na']

from pandas.api.types import CategoricalDtype

cat_type = CategoricalDtype(categories=weekdays, ordered=True)
speed['Dayweek'] = speed['Dayweek'].astype(cat_type)

speed_slow = speed[speed['Speed Limit'] <= 60 ]
speed_slow.groupby('Dayweek').count()['Crash ID'].plot(kind='bar')
```

```python
speed_medium = speed[(speed['Speed Limit'] >= 70) & (speed['Speed Limit'] < 90)]
speed_medium.groupby('Dayweek')['Crash ID'].count().plot(kind='bar')
```

```python
speed_fast = speed[speed['Speed Limit'] >= 90 ]
speed_fast.groupby('Dayweek')['Crash ID'].count().plot(kind='bar')
```

```python
speed_fast = vic_fatalaties_df[vic_fatalaties_df['Gender'] !='na']
speed_fast.groupby(by=['Dayweek','Gender'])['Crash ID'].count().plot(kind='bar')
```

```python
roads = vic_fatalaties_df[vic_fatalaties_df['National Road Type'] !='na']
roads.groupby(by=['National Road Type'])['Crash ID'].count().plot(kind='bar')#, figsize=[500,500])
```

```python
def exp_life(Life_Table, Age):
    life = Life_Table[Age]
    return life
```

```python
Life_table_males = pd.read_excel(Path("Australian_Life_Tables_2015-17_Males.xlsx"), sheet_name="Males")
Life_table_males.columns = ['Age','lx','dx','px','qx','mu','exp_life','Lx','Tx']

expected_life_males = Life_table_males['exp_life']
expected_life_males.reset_index()
#--------------------------------
Life_table_females = pd.read_excel(Path("Australian_Life_Tables_2015-17_Females.xlsx"), sheet_name="Females")
Life_table_females.columns = ['Age','lx','dx','px','qx','mu','exp_life','Lx','Tx']

expected_life_females = Life_table_females['exp_life']
expected_life_females.reset_index()
```

```python
# ages_males = vic_fatalaties_df['Age','Gender']
ages_males = vic_fatalaties_df[(vic_fatalaties_df['Gender']=='Male') & (vic_fatalaties_df['Age']!='na')]
ages_males_mean = int(ages_males['Age'].mean())
ages_males = vic_fatalaties_df[(vic_fatalaties_df['Gender']=='Male')]
ages_males['Age'] = ages_males['Age'].apply(lambda x: ages_males_mean if x=='na' else x)
ages_males.reindex()

ages_females = vic_fatalaties_df[(vic_fatalaties_df['Gender']=='Female') & (vic_fatalaties_df['Age']!='na')]
ages_females_mean = int(ages_females['Age'].mean())
ages_females = vic_fatalaties_df[(vic_fatalaties_df['Gender']=='Female')]
ages_females['Age'] = ages_females['Age'].apply(lambda x: ages_females_mean if x=='na' else x)
ages_females.reindex()

print(f'Average age of male deaths {ages_males_mean}, of female deaths {ages_females_mean}')
```

```python
ages_males['est_yrs_lost'] = ages_males['Age'].apply(lambda x: exp_life(expected_life_males,x) if x>=0 else 0)
total_male_yrs_lost = int(ages_males['est_yrs_lost'].sum())

ages_females['est_yrs_lost'] = ages_females['Age'].apply(lambda x: exp_life(expected_life_females,x) if x>=0 else 0)
total_female_yrs_lost = int(ages_females['est_yrs_lost'].sum())

print(f'Estimated male years lost {total_male_yrs_lost:,}, female {total_female_yrs_lost:,}')
```

-- Road deaths observations
1.Male driver deaths 3 times female drivers
2.Motorcycle rider deaths almost exclusively male
3 Passengers deaths are equally weighted male and female
4 Cyclists deaths are predominately male
5 Male pedestrian deaths are twice as likely as female
